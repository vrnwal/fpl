```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

import glob

import plotly.io as pio

pio.renderers.default = 'svg'
# pio.renderers.default = 'notebook'
```

```python
# sklearn imports

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
path = '../Fantasy-Premier-League-master/data/2019-20'
playerInfo = pd.read_csv(path + '/players_raw.csv')


playerInfo['Initial Cost'] = playerInfo['now_cost'] - playerInfo['cost_change_start']

playerInfo['Points v Cost'] = playerInfo['total_points'] / playerInfo['Initial Cost']

playerInfo
```

```python
playerStats = pd.read_csv(path + '/understat/understat_player.csv', encoding = "ISO-8859-1")
playerStats
```

```python
playerInfo['full_name'] = playerInfo['first_name'] + " " +playerInfo['web_name']
allInfo = playerInfo.merge(playerStats, left_on = 'full_name', right_on = 'player_name')
```

```python
### We need to split the dataset up by position, essentially, Forward, Mid, Defence, GK.
# Since the understat data does not use the players listed position, instead noting down all positions played
# in the season, we need to set up our filter

allInfo['Main Position'] = allInfo['position'].str[0]
positions = list(allInfo['Main Position'].value_counts().index)

fwdStats = allInfo[allInfo['Main Position'].isin(['F','S'])] 
midStats = allInfo[allInfo['Main Position'].isin(['M'])] 
defStats = allInfo[allInfo['Main Position'].isin(['D'])] 
gkStats = allInfo[allInfo['Main Position'].isin(['G'])] 

```

Look at xG vs Goals for strikers.

This let's us see which players have overperformed their stats, versus player's who have underperformed.


```python
fwdStats['Goal and XG Difference'] = fwdStats['goals'] - fwdStats['xG']
fig = px.scatter(fwdStats[fwdStats['goals']>=1], x="Goal and XG Difference", y="xG", color="player_name", size = 'goals', hover_name="player_name",
                title = 'xG vs Goals PL FWDs')
fig.show()
```

The plot below shows us a range of forwards, and how they performed in comparison to their expected goals.

From this we can see, players like Gabriel Jesus and Roberto Firmino severely underperformed their xG, while Danny Ings and PEA overperformed their stats, showing their prolificness in front of goal.

This is helpful because it shows us that Jesus comes across a lot of chances, which is a good sign for prospective owners, and could potentially regress to the mean.

```python
# Similarly, for midfielders, we can look at xA
midStats['Assist and XA Difference'] = midStats['assists_x'] - midStats['xA']
fig = px.scatter(midStats, x="Assist and XA Difference", y="xA", color="player_name", size = 'assists_x', hover_name="player_name",
                title = 'xA PL MIDs')
fig.show()
```

```python
midStats['Goal and XG Difference'] = midStats['goals'] - midStats['xG']
fig = px.scatter(midStats[midStats['goals']>=1], x="Goal and XG Difference", y="xG", color="player_name", size = 'goals', hover_name="player_name",
                title = 'xG vs Goals PL MIDs')
fig.show()
```

We notice that Kevin De Bruyne did obscenely well last season, and is an almost essential pick. 

To get a better idea of who else to select, we look at the database without KDB.

```python
midStats2 = midStats[midStats['player_name'] != 'Kevin De Bruyne']

midStats2['Assist and XA Difference'] = midStats2['assists_x'] - midStats2['xA']
fig = px.scatter(midStats2, x="Assist and XA Difference", y="xA", color="player_name", size = 'assists_x', hover_name="player_name",
                title = 'xA PL MIDs')
fig.show()
```

```python
midStats2['Goal and XG Difference'] = midStats2['goals'] - midStats2['xG']
fig = px.scatter(midStats2[midStats2['goals']>=1], x="Goal and XG Difference", y="xG", color="player_name", size = 'goals', hover_name="player_name",
                title = 'xG vs Goals PL MIDs')
fig.show()
```

```python
# This allows us to see how a player has performed in comparison to their cost, thus giving as an idea 
# of who to buy

top = playerInfo[playerInfo['minutes'] >= 1700].sort_values(by = 'Points v Cost', ascending = False).head(25)
bottom = playerInfo[playerInfo['minutes'] >= 1700].sort_values(by = 'Points v Cost', ascending = False).tail(25)


combinedValue = pd.concat([top, bottom])

fig = px.scatter(combinedValue, x="Initial Cost", y="Points v Cost", color="web_name", size = 'total_points', hover_name="web_name",
                title = 'Player Value')
fig.show()
```

In the above plot, we only look at players who played at least 1700 minutes, of the total 3420 in the season. 

This ensures we dont capture players who were rotation options or missed games for some reason, such as injury.

We can see that once again, Kevin De Bruyne stands out, performing exceptionally well, although his price of 10 million probably lowers how much value he brings to the team.

Burnley keeper Nick Pope is the standout performer in the League, , ending up with 170 points at a mere cost of 4.5 million.

Meanwhile, some notable disappointments were Nicolas Pepe, who had a high price tag of 9.5 million set by FPL, and was unable to live up to expectations.


An interesting thing we note is that a signicant proportion of the high value players were budget goalkeepers, in the 4, 4.5 price range, which backs the popular hypothesis, that budget keepers like Pope, Ryan and Henderson are better picks than their more expensive counterparts, like Ederson and Allison.

For someone with an idea of football, this plot also shows us that a lot of low value players are deep lying midfielders, like Guendouzi, Romeu, Kante, Fred, Winks. 

Although these players are crucial to their teams play, strictly from an FPL standpoint, they might not be the best assets, because their play is not something that is favored by the FPL point scoring system.

### Budget Players

Over the course of the 2019/20 season, several cheaper players, like Lundstram and Greenwood, performed very well, thus offering great value for money. 

It may be a good idea to look at this data to reinforce this faith in budget player options.

We define budget options by position. For our purpose, they are:

1. Goalkeepers:- 4.5 million
2. Defenders:- 5 million
3. Midfielders:- 6.5 million
4. Strikers:- 7 million

We obviously do not wanna compare pure output, because a lot of these budget player sare rotation options for their teams, they will tend to have less total points, but if they can offer a good value for their price, working as reinforcement options in Fantasy teams, their job is done.

```python
defStats['budget'] = defStats['Initial Cost'].apply(lambda x: 1 if x<= 45 else 0)
defStats
```

```python
# We look at Points Per Game for these players.
defStats['PPG'] = defStats['total_points'] / defStats['games']
fig = px.scatter(defStats, x="total_points", y="PPG", color="budget", size = defStats['PPG']**2, hover_name="web_name",
                title = 'Player Value')
fig.show()
```

```python
defGroup = defStats.groupby('Initial Cost').agg(np.median)
defGroup
```

## Modelling Analysis

We now wanna merge the two data sets so that we have all our information in one table, which will be useful later on.

We will then look at what columns might helps us predict our team for next season.

```python
allInfo.columns
```

For the correlation, we only wanna see performance statistics.

We do this because we are trying to use player performance data to predict their output for next season. Using information like their bonus points, chance of playing, and times they show up in the dream team are correlated to their performance metrics.

```python
selectInfo = allInfo[['clean_sheets', 'creativity', 'penalties_missed', 'penalties_saved',
                     'saves', 'assists_y', 'Initial Cost',  'games', 'goals',
                     'key_passes', 'npg', 'npxG',
                   'player_name', 'red_cards_y', 'shots',
                   'xA', 'xG', 'xGBuildup', 'xGChain', 'yellow_cards_y','total_points']]
playerCorr = selectInfo.corr()
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(20, 12)
sns.heatmap(playerCorr, center = 0, cmap="gist_rainbow_r", annot = True)
```

```python
train, test = train_test_split(allInfo, test_size = 0.3, random_state=42)
```

```python
def processData(data):
    
    return data[['creativity','assists_y','goals','npg','npxG','shots','xA','xG','xGBuildup','xGChain','minutes']], data['total_points']
```

```python
X_train, y_train = processData(train)
```

```python
X_train.dtypes
```

Since we are working with regression models, we define a Root Mean Square Error function which helps give us the error in our prediction.

```python
def rmse(y_actual, y_pred):
    return np.sqrt(np.mean(np.square(y_actual - y_pred)))
```

## Linear Regression Model

Linear Regression follows a simple principle of linearity between variables.

It follows the general equation of a line, 
$$y = mx + b$$.

We first train the model on the X_train, y_train which we described above. 

Initially, we test this on the train data, which is not optimal, but allows us to gain some sort of idea of our model.

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

y_pred = lr_model.predict(X_train)
```

```python
print(rmse(y_train, y_pred))
```

```python
trainLR = train.copy()
trainLR['predicted'] = y_pred
trainLR[['web_name', 'total_points','predicted']].head(15)
```

Now we test our model on the test set that we had created earlier.

```python
X_test, y_test = processData(test)
```

```python
y_test_pred = lr_model.predict(X_test)

print('RMSE for Test set:',rmse(y_test, y_test_pred))
```

```python
testLR = test.copy()
testLR['predicted'] = y_test_pred
testLR[['web_name', 'total_points','predicted']].head(15)
```
